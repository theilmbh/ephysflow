```python
logger = logging.getLogger()
handler = logging.StreamHandler()
formatter = logging.Formatter(
        '%(asctime)s %(name)-12s %(levelname)-8s %(message)s')
handler.setFormatter(formatter)
logger.addHandler(handler)
logger.setLevel(logging.INFO)
logger.debug('all modules loaded')
```

```python
% matplotlib inline

import matplotlib.pyplot as plt
matplotlib.style.use('ggplot')
```

```python
class Pst:
    def __init__(self, bird, sess, ev_name, fn=None, stim_sess=1):
        self.bird = bird
        self.sess = sess
        self.ev_name = ev_name
        self.fn = fn
        
        self.peri_pars = {'before': 30000,
                         'after': 30000}

        self.filters = {'lfp': None,
                       'spk': None,
                       'mic': None}
        
        self.evt_starts = None
        self.frames = None
        self.evt = None
        self.s_f = None
        self.ev_chunk = None
        self.stim_sess = stim_sess
        self.prb = None
        self.par = None
        self.files = {'evt': None,
                      'raw': None,
                      'spk': None}
        
        self.frames = {'chunks': None,
                      'starts': None,
                      'recs': None,
                      'rms': None,
                      'spk': None}
        
        if self.fn is None:
            self.fn = et.file_names(bird, sess)
        
        self.unit_list = None
        
        
        evt_file_path = et.file_path(self.fn, 'ss', 'sng')
        self.files['evt'] = et.open_kwe(bird, sess)        
        self.files['raw'] = et.open_kwd(bird, sess)
        self.files['spk'] = et.open_kwik(bird, sess)
        
        self.evt = evt.Song(m_name, self.files['evt'])
        self.par = et.get_parameters(bird, sess)
        self.prb = et.load_probe(bird, sess)
        
    def set_filter(self, filter_type, filter_obj):
        self.filters[filter_type] = filter_obj
        
    def get_evt_starts(self):
        if self.evt_starts is None:
            #self.evt_starts = self.evt.get_abs_start(self.files['raw'])
            self.evt_starts = self.evt.get_start()
        return self.evt_starts
    
    def get_ev_chunk(self):
        stim_id = et.stim_id(self.par, self.ev_name)
        ev_path = os.path.join(et.file_names(self.bird)['folders']['stim'], '{:03}'.format(self.stim_sess), 
                      self.par['search_motiff']['motiff_patterns'][m_name])
        self.ev_chunk = st.Chunk(st.WavData2(ev_path))
        return self.ev_chunk
    
    def get_rms(self, chan_list):
        rec_list = h5f.get_rec_list(self.files['raw'])
        rms = []
        for rec in rec_list:
            logger.info('Getting rms for rec {}'.format(rec))
            rec_data = st.H5Data(h5f.get_data_set(self.files['raw'], rec),
                    self.s_f,
                    dtype=np.float,
                    chan_list = chan_list)
            
            rms_rec = rec_data.get_rms(window_size_samples=6000, 
                              n_windows=10000, 
                              rms_func=lambda x: st.rms(self.filters['spk'].do_filter(x, axis=0)))
            rms.append(rms_rec)
        self.frames.update({'rms': np.stack(rms)}, axis=0)
        return self.frames['rms']
            
    def collect_frames(self, chan_list):
        before = self.peri_pars['before']
        span = before + self.peri_pars['after'] + self.ev_chunk.samples
        recs_of_starts = self.evt.get_rec()
        rec_list = np.unique(recs_of_starts)
        frames = []
        good_starts = []
        good_recs = []
        for rec in rec_list:
            logger.info('Collect frames for rec {}'.format(rec))
            rec_starts = self.evt_starts[recs_of_starts == rec]
            rec_data = st.H5Data(h5f.get_data_set(self.files['raw'], rec),
                                self.s_f,
                                dtype=np.float,
                                chan_list = chan_list)
            rec_frames, rec_bad = rec_data.collect_frames(rec_starts-before, span)
            frames+=(rec_frames)
            rec_good_starts = np.delete(rec_starts, rec_bad)
            good_starts.append(rec_good_starts)
            good_recs.append(np.ones_like(rec_good_starts)*rec)
        
        logger.info('Done collecting frames')
        
        self.frames.update({'chunks': frames,
                      'starts': np.concatenate(good_starts, axis=0),
                      'recs': np.concatenate(good_recs, axis=0)}
                          )
        return self.frames
    
    def get_spikes(self, thresh_factor, chan_list, min_dist=10):
        # assume frames are in place
        spikes_stack = []
        rms = self.frames['rms']
        logger.info('Getting spikes for {} frames'.format(len(self.frames['chunks'])))
        for i, fr in enumerate(self.frames['chunks']):
            rec = self.frames['recs'][i]
            logger.debug('Frame {0} - Rec {1}...'.format(i, rec))
            thresholds = thresh_factor*self.frames['rms'][rec, :]
            
            fr_filt = self.filters['spk'].do_filter(fr, axis=0)
            spk_array = tp.spikes_array(fr_filt, thresholds, min_dist=min_dist)
            spikes_stack.append(spk_array)
        self.frames.update({'spk': np.stack(spikes_stack, axis=0)})
        return self.frames['spk']
            

bird = 'z007'
sess = 'day-2016-09-11'
syn_sess = 2
m_name = 'motiff_1'

pst_sess = Pst(bird, sess, m_name)

pst_sess.set_filter('spk', 
                    fil.Filter(*fil.make_butter_bandpass(500, 10000, 30000)))
```
