```python
from __future__ import division

def sound_normalize(x):
    return (x/np.ptp(x)*60000).astype(np.int16)

def list_sum(a_list):
    s = a_list[0]
    for new_s in a_list[1:]:
        s+=new_s
    return s

def mkdir_p(path):
    try:
        os.makedirs(path)
    except OSError as exc:  # Python >2.5
        if exc.errno == errno.EEXIST and os.path.isdir(path):
            pass
        else:
            raise
            
def collect_frames(starts, span, s_f, kwd_file, recs_list, chan_list):
    frames = []

    logger.info('Collecting {} frames...'.format(starts.size))
    for i_start, start in enumerate(starts):
        if i_start % 10 == 0:
            logger.info("Frame {} ...".format(i_start))
        rec = recs_list[i_start]
        one_frame = st.Chunk(st.H5Data(kwdf.get_data_set(kwd_file, rec),
                                       s_f,
                                       dtype=np.float),
                             np.array(chan_list),
                             [start, start + span])
        frames.append(one_frame)
    return frames

def identity(x):
    return x

def collect_frames_stream(starts, span, s_f, kwd_file, recs_list, chan_list):
    recs = np.unique(recs_list)
    logger.info('Collecting {} recs...'.format(recs.size))
    all_frames = []
    for i_rec, rec in enumerate(recs):
        logger.info("Rec {} ...".format(rec))
        stream_obj = st.H5Data(kwdf.get_data_set(kwd_file, rec), s_f, dtype=np.float)
        starts_from_rec = stars[recs_list==rec]
        rec_frames = stream_obj.apply_repeated(starts, window, lambda x: x)
        all_frames.append[rec_frames]
    return all_frames


def list_apply_filter(chunk_list, filter_func, *filter_args, **filter_kwargs):
    return map(lambda x: 
               x.apply_filter(filter_func, *filter_args, **filter_kwargs), 
               chunk_list)

def get_all_bouts(bird, sess, stim_sess='001', m_name='motiff_1', dest_fold=None):
    fn = et.file_names(bird, sess)
    
    if dest_fold is None:
        dest_fold = os.path.join(fn['folders']['ss'], 'bout')
    else:
        dest_fold = os.path.join(dest_fold, bird, sess)
    
    logger.info('dest_fold : {}'.format(dest_fold))
    exp_par = et.get_parameters(bird, sess)
    fn = et.file_names(bird, sess)
    song_file_path = et.file_path(fn, 'ss', 'sng')
    song_file = h5py.File(song_file_path, 'r')

    pattern_names = exp_par['search_motiff']['motiff_patterns']
    logger.info('pattern_names : {}'.format(pattern_names))

    m_path = os.path.join(et.file_names(bird)['folders']['stim'], stim_sess, 
                          exp_par['search_motiff']['motiff_patterns'][m_name])
    m = evt.Song(m_name, song_file)
    m_chunk = st.Chunk(st.WavData2(m_path))
    s_f = m_chunk.sound.s_f
    
    kwd_file = et.open_kwd(bird, sess)
    starts = m.get_start()
    recs = m.get_rec()
    
    before = 300
    after = 1500
    audio_chans = np.array([exp_par['channel_config']['mic']])
    all_frames = collect_frames(starts-before, before+after+m_chunk.samples, 
                                s_f, 
                                kwd_file, recs, audio_chans)

    mkdir_p(dest_fold)
    for i, frame in enumerate(all_frames):
        f_name = 'bout_{0:03d}.wav'.format(i)
        f_path = os.path.join(dest_fold, f_name)
        wavfile.write(f_path, s_f, sound_normalize(frame.data.flatten()))
    return all_frames
        

def get_all_bird_bouts(bird, stim_sess='001', m_name='motiff_1', dest_fold=None):
    all_ev = et.get_bird_events(bird)
    all_sessions = np.unique(all_ev.sess[all_ev.name == m_name])
    logger.info('Found event {0} in {1} sessions'.format(m_name, all_sessions.size))
    for sess in all_sessions:
        logger.info('Getting event form sess {}'.format(sess))
        print stim_sess
        get_all_bouts(bird, sess, stim_sess=stim_sess, m_name=m_name, dest_fold=dest_fold)
    return
```
